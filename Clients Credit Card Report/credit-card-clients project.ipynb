Conver Sex to Object datatype 

```python
df['SEX'] = df['SEX'].astype('O')
```

Replace values in Sex column

```python
df['SEX'].replace(1,'male',inplace=True)
df['SEX'].replace(2,'female',inplace=True)
```

Do The same thing that We did with Sex in EDUCATION column

```python
df['EDUCATION'].value_counts()
```

```python
df['EDUCATION'].unique()
```

```python
df['EDUCATION'] = df['EDUCATION'].astype('O')
```

```python
df['EDUCATION'].replace(0,'primary',inplace=True)
df['EDUCATION'].replace(1,'perpary',inplace=True)
df['EDUCATION'].replace(2,'high_school',inplace=True)
df['EDUCATION'].replace(3,'university',inplace=True)
df['EDUCATION'].replace(4,'bacholear',inplace=True)
df['EDUCATION'].replace(5,'master',inplace=True)
df['EDUCATION'].replace(6,'PHD',inplace=True)
```

Do The same thing that We did with Sex in MARRIAE column

```python
df['MARRIAGE'] = df['MARRIAGE'].astype('O')
```

```python
df['MARRIAGE'].value_counts()
```

```python
df['MARRIAGE'].replace(2,0,inplace=True)
df['MARRIAGE'].replace(3,1,inplace=True)
```

```python
df['MARRIAGE'].value_counts()
```

```python
df['MARRIAGE'].replace(0,'single',inplace=True)
df['MARRIAGE'].replace(1,'married',inplace=True)
```

```python
df['default payment next month'] = df['default payment next month'].astype('bool')
```
