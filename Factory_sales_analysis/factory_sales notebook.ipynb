- Data Explority (EDA) and Visualization

```python
df.sample(10)
```

```python
df[{'TotalAmount','productPrice','Quantity'}].describe().round(1)
```

```python
df[{'CustomerName','CustomerCity','CustomerCountry','ContactName','CompanyName','ProductName'}].describe()
```

```python
df.groupby('CustomerName').max()['Quantity'].sort_values(ascending=False).head(5)
```

```python
df.groupby('CustomerName').max()['Quantity'].sort_values(ascending=False).head(5).plot(kind='bar',title='Top 5 Quantites per Customer',figsize=(7,5))
```

```python
df.groupby('CustomerName').max()['Quantity'].sort_values(ascending=False).tail(5).plot(kind='bar',title='Buttom 5 Quantites per Customer',figsize=(7,5))
df.groupby('CustomerName').max()['Quantity'].sort_values(ascending=False).tail(5)
```

```python
plt.figure(figsize=(20,5))
plt.title('Quantity per Date')
sn.lineplot(data=df,x='OrderDate',y='Quantity',markers=True)
```

```python
y = df.pivot_table(index='ProductName',values='productPrice',aggfunc=np.max).sort_values(by='productPrice',ascending=False).tail(5)
y
```

```python
y.plot(kind='barh',title='Product Per Price')
```

```python
x = df.pivot_table(index='ProductName',values='productPrice',aggfunc=np.max).sort_values(by='productPrice',ascending=False).head(5)
x
```

```python
x.plot(kind='barh')
```

```python
plt.figure(figsize=(20,5))
plt.title('TotalAmount Per Date')
sn.lineplot(data=df,x='OrderDate',y='TotalAmount')
plt.show()
```

```python
px.scatter(data_frame=df,x='Quantity',y='productPrice',size='Quantity',title='Quantity&price Distrbuation')
```

```python
px.box(data_frame=df,x='CompanyName',y='productPrice',title='Compaines Prices')
```

```python
px.area(data_frame=df,x='ContactName',y='Quantity',title='Contact Per Quantity')
```

```python
a = df.groupby('CustomerCity')['Quantity'].max().sort_values(ascending=False).head(5)
a
```

```python
b = df.groupby('CustomerCity')['Quantity'].max().sort_values(ascending=False).tail(5)
b
```

```python
c = df.groupby('CustomerCountry')['TotalAmount'].sum().sort_values(ascending=False).head(5)
c
```

```python
d = df.groupby('CustomerCountry')['TotalAmount'].sum().sort_values(ascending=False).tail(5)
d
```

```python
px.bar(data_frame=b,title='TOP CUSTOMER CITIES PER QUANTITY')
```

```python
px.funnel(data_frame=a,title='BUTTOM CUSTOMER CITIES PER QUANTITY')
```

```python
px.bar(data_frame=c,title='TOP CUSTOMER COUNTRIES PER TOTALS PRICES')
```

```python
d.plot(kind='pie',figsize=(10,7),title='BUTTOM CUSTOMER COUNTRIES PER TOTALS PRICES')
```
