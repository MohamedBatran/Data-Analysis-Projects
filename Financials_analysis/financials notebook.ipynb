- Import Packages

```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

- Import Data

```python
df = pd.read_excel('finincials.xlsx')
```

- Data Cleaning

```python
df['Units Sold'] = df['Units Sold'].astype('int64')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year'] = df['Year'].astype('O')
```

```python
df.groupby('Country').sum()['Units Sold'].plot(kind='barh',figsize=(7,7))
df.groupby('Country').sum()['Units Sold'].sort_values(ascending=False)
```

```python
df.plot(x='Date',y='Gross Sales',kind='area')
```

```python
df.pivot_table(index='Segment' , columns='Product' , values='Profit',aggfunc= np.sum).round(2).plot(kind='box')
```

```python
df.groupby('Month Name').sum()['Units Sold'].plot(kind='barh',title='Units Per Month',figsize=(8,7),xlabel='Units Sold')
df.groupby('Month Name').sum()['Units Sold'].sort_values(ascending=False)
```

```python
df.pivot_table(index='Product',columns='Discount Band',values=' Sales',aggfunc=np.max).round(2)
```

```python
df.plot(kind='hist',figsize=(20,10))
```

```python
sn.regplot(data=df,x='Profit',y='Sale Price')
```

```python
df.columns
```

```python
df.groupby('Year').sum()['Gross Sales'].plot(kind='pie',figsize=(8,5))
df.groupby('Year').sum()['Gross Sales'].sort_values(ascending=False)
```
